# Train set

### Pix2Pix

```
PATH = '/content/gdrive/MyDrive/VanGoghGAN_Landscapes/Dataset'
#folders = sorted(os.listdir(PATH))
folders = ['Boudin', 'Caillebotte', 'Cezanne', 'Chase', 'Drawings', 'Gauguin', 'Gorbatov', 'Guillaumin', 'Hassam', 'Korovin', 'Loiseau', 'Martin', 'Maufra', 'Metcalf', 'Monet', 'Pissarro', 'Renoir', 'Robinson', 'Rose', 'Sisley', 'Thaulow', 'Twachtman', 'Van Gogh', 'Weir']
print(folders)

for folder in folders:
  images = os.listdir(PATH+'/'+folder+'/Originals')
  for image in images:
    img = cv2.imread(PATH+'/'+folder+'/Originals/'+image)
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    inverted_gray_image = 255 - gray_image
    blurred_img = cv2.GaussianBlur(inverted_gray_image, (21,21), 0)
    inverted_blurred_img = 255 - blurred_img
    pencil_sketch_IMG = cv2.divide(gray_image, inverted_blurred_img, scale = 256.0)
    #cv2_imshow(img)
    #cv2_imshow(pencil_sketch_IMG)
    print(PATH+'/'+folder+'/Sketches/'+image[:-4]+' Sketch.jpg')
    cv2.imwrite(PATH+'/'+folder+'/Sketches/'+image[:-4]+' Sketch.jpg', pencil_sketch_IMG)
```

### CycleGAN

```
PATH = '/content/gdrive/MyDrive/VanGoghGAN_Landscapes/Dataset'
#folders = sorted(os.listdir(PATH))
folders = ['Boudin', 'Caillebotte', 'Cezanne', 'Chase', 'Drawings', 'Gauguin', 'Gorbatov', 'Guillaumin', 'Hassam', 'Korovin', 'Loiseau', 'Martin', 'Maufra', 'Metcalf', 'Monet', 'Pissarro', 'Renoir', 'Robinson', 'Rose', 'Sisley', 'Thaulow', 'Twachtman', 'Van Gogh', 'Weir']
print(folders)

for folder in folders:
  images = os.listdir(PATH+'/'+folder+'/Originals')
  for image in images:
    img = cv2.imread(PATH+'/'+folder+'/Originals/'+image)
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    inverted_gray_image = 255 - gray_image
    blurred_img = cv2.GaussianBlur(inverted_gray_image, (21,21), 0)
    inverted_blurred_img = 255 - blurred_img
    pencil_sketch_IMG = cv2.divide(gray_image, inverted_blurred_img, scale = 256.0)
    pencil_sketch_IMG_BGR = cv2.cvtColor(pencil_sketch_IMG, cv2.COLOR_GRAY2BGR)
    #cv2_imshow(img)
    #cv2_imshow(pencil_sketch_IMG_BGR)
    print(PATH+'/'+folder+'/Sketches3D/'+image[:-4]+' Sketch 3D.jpg')
    cv2.imwrite(PATH+'/'+folder+'/Sketches3D/'+image[:-4]+' Sketch 3D.jpg', pencil_sketch_IMG_BGR)
```

### Canny

```
PATH = '/content/gdrive/MyDrive/VanGoghGAN_Landscapes/Dataset'
#folders = sorted(os.listdir(PATH))
folders = ['Boudin', 'Caillebotte', 'Cezanne', 'Chase', 'Drawings', 'Gauguin', 'Gorbatov', 'Guillaumin', 'Hassam', 'Korovin', 'Loiseau', 'Martin', 'Maufra', 'Metcalf', 'Monet', 'Pissarro', 'Renoir', 'Robinson', 'Rose', 'Sisley', 'Thaulow', 'Twachtman', 'Van Gogh', 'Weir']
print(folders)

for folder in folders:
  images = os.listdir(PATH+'/'+folder+'/Originals')
  for image in images:
    img = cv2.imread(PATH+'/'+folder+'/Originals/'+image)
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    inverted_gray_image = 255 - gray_image
    blurred_img = cv2.GaussianBlur(inverted_gray_image, (21,21), 0)
    inverted_blurred_img = 255 - blurred_img
    pencil_sketch_IMG = cv2.divide(gray_image, inverted_blurred_img, scale = 256.0)
    blurred_img = cv2.GaussianBlur(pencil_sketch_IMG, (5,5), 0)
    edges = cv2.Canny(blurred_img, 100, 200)
    edges_BGR = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)
    #cv2_imshow(img)
    #cv2_imshow(edges_BGR)
    print(PATH+'/'+folder+'/Canny/'+image[:-4]+' Canny.jpg')
    cv2.imwrite(PATH+'/'+folder+'/Canny/'+image[:-4]+' Canny.jpg', edges_BGR)
```
